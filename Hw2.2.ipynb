Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
group_data = my_data.groupby(by=['Должность'])
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
position_data = my_data.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max'])
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(column, axis=1):
    return column.mean()

my_data['avg'] = my_data.groupby('Должность')['Зарплата.в.месяц'].transform(fill_avg_salary)
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
my_data['avg'].describe()
```

Збережіть отриману таблицю в CSV файл

```python
my_data.to_csv('2017_jun_final_with_avg')
```

Графіки

```python
grouped_py_data = python_data.groupby(by=['Должность']).size().reset_index(name='Кількість')
explode = [0.01, 0.01, 0.01, 0.01, 0.01]
plt.pie(
    grouped_py_data['Кількість'],
    labels=grouped_py_data['Должность'],
    shadow=False,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
    explode=explode
)
plt.title('Розподіл програмістів Python за посадами')

plt.show()
```

```python
position_data_max = position_data.sort_values(by=['max'], ascending=False).head(10)

plt.figure(figsize=(10, 6))
position_data_max.plot(kind='bar', width=1.3, color=('#772953', '#DA70D6'))
plt.xlabel('Максимальна зарплата в місяць')
plt.ylabel('Посада')
plt.title('Перші 10 посад з найвищою максимальною зарплатою')
plt.show()
```

```python
avg_data = my_data['avg'].head(32)

plt.plot(avg_data, label='Середня зарплата по посадах', linestyle='--', linewidth=2, marker='*')
plt.xlabel('Посада', fontsize='small', color='#4B0082')
plt.ylabel('Середня зарплата', fontsize='small', color='#4B0082')
plt.title('Середня зарплата по посадах', fontsize=15)
plt.legend()
plt.grid()
positions = my_data['Должность'].drop_duplicates().to_list()
plt.xticks(range(len(positions)), positions, rotation=90, fontsize=8, color='black')
plt.show()
```
